```python
import matplotlib.pyplot as plt
from PIL import Image
from pylab import *
from numpy import *

def gray(i, precision=4):
    arr = array(img)
    intervalo = (255 // precision)

    for i in range(len(arr)):
        for j in range(len(arr[i])):
            m = sum(arr[i][j]) // 3
            val = intervalo * (m // intervalo)
            arr[i][j] = [val, val, val]
    
    return Image.fromarray(arr)
```

```python
img = Image.open('./src/images/campo.jpg')

plt.subplots(2, 2, figsize=(80,80))

plt.subplot(323)
plt.imshow(gray(img, 4))

plt.subplot(324)
plt.imshow(gray(img, 8))

plt.subplot(325)
plt.imshow(gray(img, 16))

plt.subplot(326)
plt.imshow(gray(img, 32))
```
