```python
import matplotlib.pyplot as plt
from PIL import Image
from numpy import *


def substituiPixelMediaVizinhosD4(img):
    arr = array(img)
    
    for i in range(len(arr)):
        for j in range(len(arr[i])):
            arr[i][j] = mediaSeparada(getNeighborsD4(arr, i, j))
    return Image.fromarray(arr) 


def substituiPixelMediaVizinhosD8(img):
    arr = array(img)
    
    for i in range(len(arr)):
        for j in range(len(arr[i])):
            arr[i][j] = mediaSeparada(getNeighborsD8(arr, i, j))
    return Image.fromarray(arr)
                 
            
def getNeighborsD4(arr, i, j):
    neighborhood = [                  
        getPixel(arr, i -1, j), 
        getPixel(arr, i + 1, j),
        getPixel(arr, i, j -1),
        getPixel(arr, i, j+1)
    ]
        
    return list(filter(lambda x: x is not None, neighborhood))    


def getNeighborsD8(arr, i, j):

    neighborhood = [ 
        getPixel(arr, i -1, j), 
        getPixel(arr, i + 1, j),
        getPixel(arr, i, j -1),
        getPixel(arr, i, j+1), 
        getPixel(arr, i-1, j-1),
        getPixel(arr, i-1, j+1),
        getPixel(arr, i+1, j-1),
        getPixel(arr, i+1, j+1)
    ]
    
    return list(filter(lambda x: x is not None, neighborhood))    
            
            
def mediaRGB(arr):
    a,b,c = 0,0,0
    for i in range(len(arr)):
        a += arr[i][0]
        b += arr[i][1]
        c += arr[i][2]
    
    return [a//len(arr), b//len(arr), c//len(arr)]
            

def getPixel(im, i, j):
    if i < 0 or i > len(im) -1:
        return None
    elif j < 0 or j > len(im[0])-1:
        return None
    else:
        return im[i][j]
     
```

```python
img = Image.open('./src/images/Lenna.png')
plt.subplots(1, 3,  figsize=(25, 25))
plt.rcParams.update({'font.size':20})

plt.subplot(131)
plt.imshow(img)

plt.subplot(132)
plt.imshow(substituiPixelMediaVizinhosD4(img))

plt.subplot(133)
plt.imshow(substituiPixelMediaVizinhosD8(img))
```
