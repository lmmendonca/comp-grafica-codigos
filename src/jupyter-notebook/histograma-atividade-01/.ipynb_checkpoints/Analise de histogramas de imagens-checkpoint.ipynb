```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from pylab import * 


def geraHistoSeparado(path):
    img = np.array(Image.open(path))
    imgC = Image.open(path).convert('L')
    imgCarray = np.array(imgC)

    red = img[:, :, 0]
    green = img[:, :, 1]
    blue = img[:, :, 2]

    plt.subplot(231)
    plt.imshow(Image.fromarray(img))
    plt.subplot(232)
    plt.imshow(imgC)
    plt.subplot(233)
    plt.hist(np.ravel(red), bins=20, color='r', label='red')
    plt.subplot(234)
    plt.hist(np.ravel(green), bins=20, color='g', label='green')
    plt.subplot(235)
    plt.hist(np.ravel(blue), bins=20, color='b', label='blue')
    plt.subplot(236)
    plt.hist(np.ravel(imgCarray), bins=20, color='gray', label='gray')
```

```python
geraHistoSeparado("../images/blue.jpeg")
```

```python
geraHistoSeparado("../images/green.jpeg")
```

```python
geraHistoSeparado("../images/red.jpg")
```
