```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from pylab import * 

def grayScale(img):
    a = array(img)
    for i in range(len(a)):
        for j in range(len(a[i])):
            t = sum(a[i][j]) // 3
            a[i][j] = [t, t, t]
    return Image.fromarray(a)

def geraHistoSeparado(path):
    img = np.array(Image.open(path))
    imgC = grayScale(Image.open(path))
    imgCarray = np.array(imgC)

    red = img[:, :, 0]
    green = img[:, :, 1]
    blue = img[:, :, 2]

    plt.subplot(231)
    plt.imshow(Image.fromarray(img))
    plt.subplot(232)
    plt.imshow(imgC)
    plt.subplot(233)
    plt.hist(np.ravel(red), bins=20, color='r', label='red')
    plt.subplot(234)
    plt.hist(np.ravel(green), bins=20, color='g', label='green')
    plt.subplot(235)
    plt.hist(np.ravel(blue), bins=20, color='b', label='blue')
    plt.subplot(236)
    plt.hist(np.ravel(imgCarray), bins=20, color='gray', label='gray')
```

```python
geraHistoSeparado("./src/images/blue.jpeg")
```

```python
geraHistoSeparado("./src/images/green.jpeg")
```

```python
geraHistoSeparado("./src/images/red.jpg")
```
