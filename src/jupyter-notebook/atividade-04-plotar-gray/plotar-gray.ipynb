```python
import matplotlib.pyplot as plt
from PIL import Image
from pylab import *
from numpy import *

def gray(i, precision=4):
    arr = array(img)
    intervalo = (255 // precision)

    for i in range(len(arr)):
        for j in range(len(arr[i])):
            m = sum(arr[i][j]) // 3
            val = intervalo * (m // intervalo)
            arr[i][j] = [val, val, val]
    
    return Image.fromarray(arr)

```

```python
img = Image.open('./src/images/campo.jpg')
plt.subplots(4, 2, figsize=(80,80))
plt.rcParams.update({'font.size':50})

plt.subplot(421)
plt.imshow(gray(img, 4))

plt.subplot(422)
plt.hist(array(gray(img, 4)).ravel())

plt.subplot(423)
plt.imshow(gray(img, 8))

plt.subplot(424)
plt.hist(array(gray(img, 8)).ravel())


plt.subplot(425)
plt.imshow(gray(img, 16))

plt.subplot(426)
plt.hist(array(gray(img, 16)).ravel())

plt.subplot(427)
plt.imshow(gray(img, 32))

plt.subplot(428)
plt.hist(array(gray(img, 32)).ravel())
```
